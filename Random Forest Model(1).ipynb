```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV,KFold
```

```python
y=df3.pop('Survived')
```

```python
x=df3
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=9)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
#now with hyperparametr tuning building a model...
```

```python
n_estimators = [10, 100, 1000, 2000]
max_depth = [None, 5, 10, 20]
min_samples_leaf=[1,3,5,7]
bootstrap=[True,False]
param_grid = dict(n_estimators=n_estimators, max_depth=max_depth,min_samples_leaf=min_samples_leaf,bootstrap=bootstrap)
```

```python
rf=RandomForestClassifier(random_state=9)
grid=GridSearchCV(estimator=rf,
                 param_grid=param_grid,
                 cv=3,
                 verbose=2,
                 n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
```

```python
y_pred_best=grid.predict(x_test)
```

```python
print(accuracy_score(y_pred_best,y_test))
```

```python
print(classification_report(y_pred_best,y_test))
```

```python
print(confusion_matrix(y_pred_best,y_test))
```

```python
grid.best_score_
```
